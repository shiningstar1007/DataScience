```python
import pandas as pd

#encding="cp949" or "euc-kr"
#메소드 help 보는 법 : Tap 또는 Shift + Tap
#csv file Load
csvName1="주택도시보증공사_전국 평균 분양가격(2019년 12월).csv"
csvName2="전국 평균 분양가격(2013년 9월부터 2015년 8월까지).csv"

data_Last = pd.read_csv(csvName1)
data_First = pd.read_csv(csvName2)
```

```python
#데이터 타입 변경

data_Last["분양가격"] = pd.to_numeric(data_Last["분양가격(㎡)"], errors='coerce')
data_Last["분양가격"]
```

```python
#평당 분양가격 구하기

data_Last["평당분양가격"] = data_Last["분양가격"] * 3.3
data_Last
```

```python
# 분양가격 요약

data_Last["분양가격"].describe(include='all')
```

```python
#규모구분을 전용면적으로 변경

data_Last["전용면적"] = data_Last["규모구분"].str.replace("전용면적", "")
data_Last["전용면적"] = data_Last["전용면적"].str.replace("초과", "~")
data_Last["전용면적"] = data_Last["전용면적"].str.replace("이하", "")
data_Last["전용면적"] = data_Last["전용면적"].str.replace(" ", "").str.strip()
data_Last["전용면적"]
```

```python
# 필요없는 컬럼 제거
# data_Last.drop(["규모구분", "분양가격(㎡)"], axis=1)
data_Last = data_Last.drop(columns=["규모구분", "분양가격(㎡)"])
```

```python
# melt 사용하여 2개의 데이터셋을 같은 형태로 만들기

data_First_melt = data_First.melt(id_vars="지역", var_name="기간", value_name="평당분양가격")
data_First_melt.columns = ['지역명', '기간', '평당분양가격']
data_First_melt
```

```python
#split 사용해서 연도와 월을 분리

def parse_year(date):
    year = date.split("년")[0]
    return int(year)

def parse_month(date):
    month = date.split("년")[1].replace("월", "")
    return int(month)


data_First_melt["연도"] = data_First_melt["기간"].apply(parse_year)
data_First_melt["월"] = data_First_melt["기간"].apply(parse_month)
data_First_melt.head(1)
```

```python
data_Last.columns.to_list()
```

```python
cols = ['지역명', '연도', '월', '평당분양가격']
```

```python
data_Last_prepare = data_Last.loc[data_Last["전용면적"] == "전체", cols].copy()
data_Last_prepare 
```

```python
data_First_prepare = data_First_melt[cols].copy()
data_First_prepare
```

```python
#concat으로 2개의 데이터프레임 합치기

data_concat = pd.concat([data_First_prepare, data_Last_prepare])
data_concat["연도"].value_counts(sort=False)
```

```python
#pivot_table 사용하기
#연도를 인덱스로, 지역명을 컬럼으로 평당 분양가격

data_heatmap = data_concat.pivot_table(index="연도", columns="지역명", values="평당분양가격").round()
```

```python
#피봇 테이블을 히트맵으로 변경

plt.figure(figsize=(15, 7))
sns.heatmap(data_heatmap, cmap="Blues", annot=True, fmt=".0f")
```

```python
# transpose == T (행과열을 바꿔줌)
plt.figure(figsize=(15, 7))
sns.heatmap(data_heatmap.T, cmap="Blues", annot=True, fmt=".0f")
```

```python
# Groupby로 연도와 지역명으로 평당 분양가격
data_heatmap = data_concat.groupby(["연도","지역명"])["평당분양가격"].mean().unstack().round()
```
