```python
import pandas as pd

#encding="cp949" or "euc-kr"
#메소드 help 보는 법 : Tap 또는 Shift + Tap
#csv file Load
csvName1="주택도시보증공사_전국 평균 분양가격(2019년 12월).csv"
csvName2="전국 평균 분양가격(2013년 9월부터 2015년 8월까지).csv"

data_Last = pd.read_csv(csvName1)
data_Frist = pd.read_csv(csvName2)
```

```python
#결측치 보기

data_Last.isnull()
```

```python
#결측치 개수

data_Last.isnull().sum()
```

```python
#데이터 타입 변경

data_Last["분양가격"] = pd.to_numeric(data_Last["분양가격(㎡)"], errors='coerce')
data_Last["분양가격"]
```

```python
#평당 분양가격 구하기

data_Last["평당분양가격"] = data_Last["분양가격"] * 3.3
data_Last
```

```python
# 분양가격(㎡) 요약

data_Last["분양가격(㎡)"].describe()
```

```python
# 분양가격 요약

data_Last["분양가격"].describe(include='all')
```

```python
#규모구분을 전용면적으로 변경

data_Last["전용면적"] = data_Last["규모구분"].str.replace("전용면적", "")
data_Last["전용면적"] = data_Last["전용면적"].str.replace("초과", "~")
data_Last["전용면적"] = data_Last["전용면적"].str.replace("이하", "")
data_Last["전용면적"] = data_Last["전용면적"].str.replace(" ", "").str.strip()
data_Last["전용면적"]
```

```python
# 필요없는 컬럼 제거
# data_Last.drop(["규모구분", "분양가격(㎡)"], axis=1)
data_Last = data_Last.drop(columns=["규모구분", "분양가격(㎡)"])
```

```python
# groupby는 series 형태, 속도면에서 조금 더 빠름
#groupby로 지역별 평당분양가격 데이터 집계 series 형태
#help -> groupby(["컬럼명"])[계산할 컬럼명].연산메소드()

data_Last.groupby(["지역명"])["평당분양가격"].describe()
```

```python
#groupby로 전용면적으로 평당분양가격 평균값

data_Last.groupby(["지역명","전용면적"])["평당분양가격"].describe()
```

```python
#groupby로 지역명, 전용면적으로 평당분양가격 평균값뒤에 소수점 제거
data_Last.groupby(["지역명","전용면적"])["평당분양가격"].mean().unstack().round()
```

```python
#pivot_table -> DataFrame 형태

data_Last.pivot_table(index=["지역명"], values=["평당분양가격"], aggfunc="mean")
```

```python
data_Last.groupby(["전용면적"])["평당분양가격"].mean()
```

```python
# data_Last.groupby(["전용면적"])["평당분양가격"].mean()
# pd.pivot_table 하고 pd.pivot의 차이점은 pd.pivot은 연산은 하지 않고 데이터 형태만 바꿈 (aggfunc이 없음)

data_Last.pivot_table(index="전용면적", values="평당분양가격")
```

```python
data_Last.groupby(["지역명","전용면적"])["평당분양가격"].describe()
```
